# Data Cleaning

```python
df_auto.drop_duplicates(keep = 'first', inplace = True, ignore_index = False)
df_auto.head()
```

```python
df_auto.info()
```

```python
df_auto = df_auto[~df_auto.Date.str.contains("днес", na=False)]
df_auto.info()
```

```python
price_lv = []
for index, row in df_auto.iterrows():
    string = row['Price']
    if 'лв' in string:
        price = ''.join(re.findall(r'\d+', string))
        price_lv.append(np.float64(price))
    elif 'EUR' in string:
        price = ''.join(re.findall(r'\d+', string))
        price = np.float64(price) * 1.95
        price_lv.append(price)
    elif 'USD' in string:
        price = ''.join(re.findall(r'\d+', string))
        price = np.float64(price) * 1.90
        price_lv.append(price)
    else:
        price_lv.append(np.nan)
    
df_auto['Price BGN'] = price_lv   
```

```python
del df_auto['Price']
df_auto.head()
```

```python
df_auto.rename(columns={'Year':'Year Manifacture', 'Engine_type':'Engine Type',
                       'Date': 'Year Post'}, inplace=True)
df_auto.head()
```

```python
# extract years from "Year Manifacture" 
only_year = []
for index, row in df_auto.iterrows():
    string = row['Year Manifacture']
    string = ''.join(re.findall("\d+", string))
    only_year.append(string)
```

```python
df_auto["Year Manifacture"] = only_year
```

```python
df_auto.head()
```

```python
# replace blank spaces with np.nan
df_auto = df_auto.replace(r'^\s*$', np.nan, regex=True)
df_auto.head()
```

```python
df_auto['Year Post'] = only_year
df_auto.head()
```

```python
df_auto['Year Post_new'] = df_auto['Year Post'].apply(lambda row: f'12/31/{row}')
```
